```python
# import package and its set of stopwords
from wordcloud import WordCloud, STOPWORDS as stopwords
```

```python
hp_text = open("Book 1 - The Philosopher's Stone.txt", 'rt', encoding = 'UTF-8').read()
hp_text
```

```python
hpwc = WordCloud(background_color = 'white', 
                collocations = False,
                stopwords = stopwords)
hpwc.generate(hp_text)
```

```python
hpwc.words_
```

```python
from nltk.corpus import stopwords
nltk.download('stopwords')
stopwords_ch = stopwords.words('english')
newStopWords = ['s', 't', 'J', 'd', 'didn', 've', 'don', 'll', 're', 'm',
               'will', 'Page']
stopwords_ch.extend(newStopWords)
```

```python
hpwc = WordCloud(background_color = 'white', 
                collocations = False,
                stopwords = stopwords_ch)
hpwc.generate(hp_text)
```

```python
hpwc.words_
```
